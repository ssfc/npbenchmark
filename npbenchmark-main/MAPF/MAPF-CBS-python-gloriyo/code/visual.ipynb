```python
generated = {'CBS + Standard': [300, 320, 330, 346, 370],
             'CBS + Disjoint': [280, 275, 250, 233, 200],
             'ICBS + Standard': [208, 206, 204, 202, 200],
             'ICBS + Disjoint': [180, 175, 166, 150, 130],
            } 

df = pd.DataFrame(generated, index=['instance_1','instance_2','instance_3','instance_4','instance_5'])

print(df)
```

```python
sns.set_style("whitegrid")
p = sns.lineplot(data=df, markers=True, palette='rocket_r')
p.set_xlabel("Instances (custom)", fontsize=14)
p.set_ylabel("Number of nodes generated", fontsize=14)
```

```python
# Replace with real numbers from batch test as lists with 50 elements, for 50 batch tests
batch_generated = {'CBS + Standard': np.random.randint(700, 1200, size=50),
                'CBS + Disjoint': np.random.randint(500, 1250, size=50),
                'ICBS + Standard': np.random.randint(570, 650, size=50),
                'ICBS + Disjoint': np.random.randint(300, 750, size=50),
                } 
df_batch_gen = pd.DataFrame(batch_generated)
print(df_batch_gen)
```

```python
# Replace with real numbers from batch test as lists with 50 elements, for 50 batch tests
batch_expanded = {'CBS + Standard': np.random.randint(354, 800, size=50),
                'CBS + Disjoint': np.random.randint(300, 750, size=50),
                'ICBS + Standard': np.random.randint(280, 525, size=50),
                'ICBS + Disjoint': np.random.randint(150, 375, size=50),
                } 
df_batch_exp = pd.DataFrame(batch_expanded)
print(df_batch_exp)
```

```python
# For the batch

sns.set_style("whitegrid")
fig, ax = plt.subplots()

# Plot nodes generated
# sns.set_color_codes('pastel')
sns.barplot(data=df_batch_gen,
            label='Nodes generated', color='lightblue')

# Plot nodes expanded
# sns.set_color_codes('muted')
sns.barplot(data=df_batch_exp,
            label='Nodes expanded', color='black')
ax.set_ylabel('')

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlabel='Average nodes generated and nodes expanded: batch test')
sns.despine(left=True, bottom=True)
# plt.savefig('ten_most_reviewed.png', bbox_inches='tight')
```
